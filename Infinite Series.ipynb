Write the first five terms of the recursively defined sequence $$a_1 = 3, a_{k+1} = 2(a_k -1)$$

```python
## the next value is this value -1 and times 2

a=0
for i in np.arange(3,8,1):
    a += 2*(i-1)
    print(a)
```

Write the first five terms of the recursively defined sequence $$a_1 = 6, a_{k+1}= \frac{1}{3}a_k^2$$

```python
#The next value is 1/3rd the square of this value.
a = 0
for i in np.arange(6,11,1):
    a += (1/3)*i**2
    print(a)
```
