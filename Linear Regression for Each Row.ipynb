```python
import numpy as np
import pandas as pd
from scipy.stats import linregress

np.random.seed(1997)
df = pd.DataFrame(pd.np.random.rand(100, 10))

df['Intercept']= df.apply(lambda x: linregress(range(0,len(df.columns)), x).intercept,axis=1)
df['Slope'] = df.apply(lambda x: linregress(range(0,len(df.columns)), x).slope,axis=1)
```
