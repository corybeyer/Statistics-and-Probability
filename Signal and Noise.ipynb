# Signal and Noise

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

# Set the color palette
sns.set_palette(sns.color_palette("bone"))
sns.set_style("darkgrid")
```

```python
x = np.arange(0,25,.1)
y = np.sin(x)

noise = np.random.normal(0,1,250)
y_ = y + noise
```

```python
fig,ax = plt.subplots(3,1, figsize=(10,9))
fig.tight_layout()
sns.lineplot(x=x,y=y, ax=ax[2], label='Signal')
sns.lineplot(x=x, y=y, color='red',ax=ax[1], label = 'Signal')
sns.lineplot(x=x,y=noise, ax=ax[1], label='Noise')
sns.lineplot(x=x, y=y_ , ax= ax[0], label='Signal + Noise')
```
